__Outlier replacement__: <br> Used the __interquartile range and median__ approach for detecting the outliers i.e. considering any value  1.5*IQR below the first quartile or above the third quartile as outliers. Followed by replacing the lower values and higher values 1 and 100 

In the below outlier_detect funtion, first i have passed all the columns to find Q1,Q3 values, i.e. 

Q1  = "middle" value in the first half of the selected column <br> 
Q3  = "middle" value in the second half of the selected column<br> 
IQR =  Interquartile range is equal to Q3 minus Q1<br> 
x   =  Created numpy array of the each column of dataset <br>

Now each value of the column is being passed through loop. If the value is less than the LTV, the value is repalced by 1, while value is greater than the UTV, the value has been replace by 100.So all the value above and below the UTV, LTV are being replaced by 100 and 1 respectively.

```python
def outlier_detect(Df):
    for i in Df.describe().columns:
        Q1=Df.describe().at['25%',i]
        Q3=Df.describe().at['75%',i]
        IQR=Q3 - Q1
        LTV=Q1 - 1.5 * IQR
        UTV=Q3 + 1.5 * IQR
        x=np.array(Df[i])
        p=[]
        count1=0
        count2=0
        for j in x:
            if j < LTV:
                p.append(1)
                count1= count1+1
            elif j>UTV:
                p.append(100)
                count2= count2+1
            else:
                p.append(j)
        Df[i]=p
        print(i,LTV,UTV,count1,count2)
    return Df
```

Using the MinMaxScaler approach transforms the each feature values range (1,100) 

# Scaling Function:

J_max = maximum value in each column <br>
J_min = minimum value in each column <br>

Now each value of the particular column has been passed thorugh the loop. If the value is 1 and 100 i.e. outlier, as the values are already scaled, they are not changed. while other values are being scaled using the below formula.

p = 1+(99*(j-J_min))/(J_max - J_min))
 
now all the values in the dataset are being scaled between 1 and 100.

```python
def Scaling(Df):
    for i in Df.describe().columns:
        J_max=Df.describe().at['max',i]
        J_min=Df.describe().at['min',i]
        x=np.array(Df[i])
        p=[]
        count1=0
        count2=0
        for j in x:
            if j == 1:
                p.append(1)
                count1= count1+1
            elif j == 100:
                p.append(100)
                count2= count2+1
            else:
                p.append(1+(99*(j-J_min))/(J_max - J_min))
        Df[i]=p
        print(i,count1,count2)
    return Df
```

```python
Scaling(df)
df = df.round(0)
df.head()
# df.to_csv('ECS Scaled_Values.csv')
```
